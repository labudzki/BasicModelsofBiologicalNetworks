```python
import pysces
```

```python
model_description = '''

r1:
{2}p = p2
k1f * p**2 - k1b * p2

#finish the other reactions

#parameters
k1f = 1
k1b = 1
k2f = 1
k2b = 1
k3f = 1
k3b = 1
k4f = 1
k4b = 1
k5f = 1
k5b = 1

#initial concentrations
p = 1
s = 10
p2 = 0
sp2 = 0
p2s = 0
sp2s = 0

'''
```

```python
model = pysces.model('my_model', loader='string', fString=model_description)
```

```python
model.doSim(end=2, points = 10000)
```

```python
model.SimPlot()
```

```python
# Finding the steady state values
model.doStateShow()
```
